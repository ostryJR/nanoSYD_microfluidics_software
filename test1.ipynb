```python
import serial
import time

ser = serial.Serial(port='COM3', baudrate=9600, bytesize=8, parity='N', stopbits=1, timeout=1)

def decompose_reply(reply):
    print(f"Decomposing reply: {reply}")
    address = int(reply[1:2].decode())
    status = str(chr(reply[2]))
    data = str(reply[3:].decode(errors="ignore"))[:-3]
    return address, status, data

def send_command(ser, cmd):
    print(f"Sending command: {cmd}")
    ser.write(cmd)
    time.sleep(0.5) # wait for the response
    reply = ser.read_all()
    print(f"Received reply: {decompose_reply(reply)}")
    print(f"Raw reply: {reply}")
    #print(f"Hex reply: {hex(reply.decode())}")
#    if reply==b'/0`\x03\r\n':
#        print(f"[GOOD]Received reply: {reply}")
#    elif reply==b'/0@\x03\r\n':
#        print(f"[WORK]Received reply: {reply}")
#    else:
#        print(f"[ERROR]Received reply: {reply}")
#    return reply

time.sleep(2) # wait for the serial connection to initialize

print(f"Serial connection established: {ser.is_open}, on port: {ser.port}")

#cmd = b"/1ZR\r" # inicialize pump 1
#print(f"Sending command: {cmd}")
#ser.write(cmd)

#time.sleep(0.5) # wait for the response

#reply = ser.read_all()
#print(f"Received reply: {reply}")

# /1 pump 1
# /2 pump 2

send_command(ser, b"/1ZR\r") # initialize pump 1

print("Pump initialized \n")
time.sleep(5)
#send_command(ser, b"/1A24000R\r")
send_command(ser, b"/1P1500R\r")
print("Dispense command sent \n")
time.sleep(1)
#send_command(ser, b"/1Q\r")
send_command(ser, b"/1?0\r")

time.sleep(5)
send_command(ser, b"/1?0\r")

time.sleep(5)
send_command(ser, b"/1?0\r")
ser.close()
```
